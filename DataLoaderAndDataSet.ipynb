**Dataset** class fetch the data form the memory It use three method
. __init()__ it is used how to load or from where to get data
.__len()__it is very simple it is used to find the lenght of the sample or the batch size of the data
.__getitem()__it is used to return the particular row using the index
**DataLoader** it fetches the row form the dataset class and sufflle its indixes and make the batches of the indixes and call the dataset method.__get__item() and using this method ot convert the indixezs into the rows

```python
#Step 1 import the necessary libraries
from sklearn.datasets import make_classification
import torch
```

```python
#Step Create a synthetic classification dataset using sklearn
X,y=make_classification(
    n_samples=1000,#no of the samples
    n_features=10, #no of the features
    n_classes=2, #no of the class
    n_redundant=0, #no of the redundant features
    n_informative=2,#no of the informative feature
    random_state=42# for reproductivitiy
    )
```

```python
print(X.shape)
print(y.shape)
```

```python
#firsr we have to convert the arrays into tensor
x=torch.tensor(X,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.long)
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    #it set the
    self.features=features
    self.labels=labels;
  def __len__(self):
    return self.features.shape[0]
  def __getitem__(self,index):
    return self.features[index],self.labels[index]
```

```python
dataset=CustomDataset(x,y)
```

```python
print(len(dataset))
print(dataset[0])
```

What is the magic function
Magic Function is also called the dunder method in the python which is automatically called by the class

```python
#Now we are going to start the data loader
dataloader=DataLoader(dataset,batch_size=32,shuffle=True)
```

```python
for batch_features , batch_labels in dataloader:
  print(batch_features.shape)
  print(batch_labels.shape)
```
