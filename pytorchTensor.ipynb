```python
a=torch.randint(size=(2,3),low=0,high=10)
b=torch.randint(size=(2,3),low=10,high=20)
print(a)
print(b)
```

```python
#mathematiacal operation
print(a+2)
print(a-2)
print(a*2)
print(a/3)
print((a*100)//3)
print(((a*100)//3)%2)
print(x**2)
```

```python
#Element wise operation
print(a+b)
print(a-b)
print(a/2)
print(a%b)
print(a//b)
print(a**b)
```

```python
#more operation
d=torch.linspace(-2,4,10)
d
```

```python
print(torch.abs(d))
print(torch.round(d))
print(torch.negative(d))
print(torch.ceil(d))
print(torch.floor(d))
print(torch.clamp(d,min=2,max=3))
```

```python
f=torch.randint(size=(2,3),low=0,high=10,dtype=torch.float64)
f
```

```python
print(torch.sum(f))
print(torch.sum(f,dim=0))
print(torch.sum(f,dim=1))
print(torch.mean(f))
print(torch.mean(f,dim=0))
print(torch.mean(f,dim=1))
print(torch.max(f))
print(torch.min(f))
print(torch.median(f))
print(torch.prod(f))
print(torch.argmax(f))
print(torch.argmin(f))
print(torch.var(f))
print(torch.std(f))
```

```python
#matrix3 operation
print(torch.multiply(a,b))
x=torch.tensor([1,2])
y=torch.tensor([3,4])
print(torch.dot(x,y))
```

```python
#transpose
print(torch.transpose(f,0,1))
```

```python
h=torch.rand(2,2)
print(torch.det(h))
print(torch.inverse(h))
```

```python
#comparision operator
print(a>b)
print(a<b)
print(a==b)
print(a!=b)
```

```python
#sum special function
print(torch.softmax(f,dim=0))
print(torch.log(f))
print(torch.exp(f))
print(torch.sigmoid(f))
print(torch.relu(f))
```

```python
#inpalce operation
print(a.add_(b))
print(a.relu_())
```

```python
#Copy function
t=a.clone()
print(t)
```

```python
#tensor operation on the cuda
device=torch.device('cuda')
#creating a new tensor on the gpu
print(torch.rand((2,3),device=device))
#Moving the existing tensor to the GPU
s=f.to(device)
print(s+5)
```

```python
#reshape
h=torch.ones(4,4)
print(h.reshape(2,2,2,2))
print(h.flatten())
```

```python
#permute
g=torch.rand(2,3,4)
print(g)
#apply the permute
print(g.permute(2,0,1))
```

```python
#unsqueeze and squeeze
k=torch.rand(226,226,3)
print(k.unsqueeze_(0).shape)
#squueze
d=torch.rand(0,20)
print(d.squeeze_(0).shape)
```
