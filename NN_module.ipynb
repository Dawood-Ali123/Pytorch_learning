# After updation the code we  are going to use the sequential **container**

```python
class Model(nn.Module):
    def __init__(self,num_features):
      super().__init__();#here we call the constructor of the parent class
      self.network=nn.Sequential(
      nn.Linear(num_features,3),
      nn.ReLU(),
      nn.Linear(3,1),
      nn.Sigmoid()
      )
    def forward(self,x):
      #here we have to olny pass the container
      x=self.network(x)
      return x
```

```python
features=torch.rand(10,5)
model=Model(features.shape[1])
print(model(features))
```

```python
from torchinfo import summary
summary(model,input_size=(10,5))
```
